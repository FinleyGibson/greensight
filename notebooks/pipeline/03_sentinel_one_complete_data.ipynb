```python
from greensight.utils import DIR_DATA
import pandas as pd
import numpy as np
import json
```

```python
# config
green_belt = "Merseyside and Greater Manchester"
```

```python
# load DataFrame
s1_path = DIR_DATA / "processed_data/sentinel_one_df.hdf"

s1_data = pd.read_hdf(s1_path, key="df")
s1_data.head(5)
```

```python
# load weights DataFrame
weights_path = DIR_DATA / "processed_data/sentinel_one_weights.hdf"

df_weights = pd.read_hdf(weights_path, key="df")
df_weights.head(5)
```

```python
# weights df shape must equal data shape
assert df_weights.shape == s1_data.shape
```

```python
# combine
s1_combined = s1_data*df_weights
s1_combined = s1_combined.groupby(level=["greenbelt", "band"], axis=1).sum()
s1_combined.head(5)
```

```python
# load JULES data
path_soil = DIR_DATA / "processed_data/jules_soil.hdf"
path_veg = DIR_DATA / "processed_data/jules_veg.hdf"
```

```python
jules_soil = pd.read_hdf(path_soil, key="df")
jules_soil = jules_soil.resample("MS").mean()
jules_soil.columns.name = "greenbelt"
jules_soil.head(5)
```

```python
jules_veg = pd.read_hdf(path_veg, key="df")
jules_veg = jules_veg.resample("MS").mean()
jules_veg.columns.name = "greenbelt"
jules_veg.head(5)
```

```python
jules_soil.columns = pd.MultiIndex.from_product([jules_soil.columns, ["soil"]], names=["greenbelt", "band"])
jules_soil.head(5)
```

```python

jules_veg.columns = pd.MultiIndex.from_product([jules_veg.columns, ["vegetation"]], names=["greenbelt", "band"])
jules_veg.head(5)
```

```python
print(jules_soil.shape)
print(jules_veg.shape)
print(s1_combined.shape)
```

```python
print(max(s1_combined.index))
print(max(jules_soil.index))
print(max(jules_veg.index))
expected_rows = s1_combined.shape[0] - (max(s1_combined.index).month - max(jules_soil.index).month)

expected_rows
```

```python
expected_cols = s1_combined.shape[1] + jules_soil.shape[1] + jules_veg.shape[1]
expected_cols
```

```python
df_all = pd.concat([jules_soil, jules_veg, s1_combined], axis=1, join="inner")
df_all = df_all.groupby(level=0, axis=1).apply(
    lambda df: df.droplevel(0, axis=1).T.drop_duplicates().T
)
print(df_all.shape)
df_all.head(5)
```

```python
assert df_all.shape == (expected_rows, expected_cols)
```

```python
path_out = DIR_DATA / "processed_data/sentinel_one_all.hdf"
df_all.to_hdf(path_out, key="df", mode="w", format="table")
```
