```python
import pandas as pd
from greensight.utils import DIR_DATA
from datetime import datetime
```

```python
# load JULES data
path_soil_carbon = DIR_DATA / "JULES/TW_UK_ERA5_BASE_V4_greenbelt16_ts_cs_gb_interp_final__v2.csv"
path_veg_carbon = DIR_DATA / "JULES/TW_UK_ERA5_BASE_V4_greenbelt16_ts_cv_interp_final__v2.csv"
assert path_soil_carbon.is_file()
assert path_veg_carbon.is_file()

df_soil_raw = pd.read_csv(path_soil_carbon)
df_veg_raw = pd.read_csv(path_veg_carbon)
```

```python
df_soil = df_soil_raw.copy()
df_soil.index = pd.to_datetime(df_soil["YrMthDay"].astype(str), format="%Y%m%d")
df_soil.index.name = "date"
df_soil.head(3)
```

```python
df_soil = df_soil.drop(columns=["YrMthDay", "year", "mth_nr", "day_of_month", "MthDay", "Yr_mth_nr", "mth"])
df_soil.head(3)
```

```python
df_soil.columns = [c[:-25].replace("_", " ") for c in df_soil.columns]
df_soil.head(5)
```

```python
# vegetation
df_veg_raw.head(5)
```

```python

df_veg = df_veg_raw.copy()
df_veg.index = pd.to_datetime(df_veg["YrMthDay"].astype(str), format="%Y%m%d")
df_veg.index.name = "date"
df_veg.head(3)
```

```python
df_veg = df_veg.drop(columns=["YrMthDay", "year", "mth_nr", "day_of_month", "MthDay", "Yr_mth_nr", "mth"])
df_veg.head(5)
```

```python
df_veg.columns = [c[:-22].replace("_", " ") for c in df_veg.columns]
df_veg.head(5)
type(df_veg.index[0])
```

```python
path_soil_out = DIR_DATA / "processed_data/jules_soil.hdf"
path_veg_out = DIR_DATA / "processed_data/jules_veg.hdf"
```

```python
df_soil.to_hdf(path_soil_out, key="df", mode="w", format="table")
df_veg.to_hdf(path_veg_out, key="df", mode="w", format="table")
```

```python
df_soil_r = pd.read_hdf(path_soil_out, key="df")
df_soil_r.head(5)
```

```python

df_veg_r = pd.read_hdf(path_veg_out, key="df")
df_veg_r.head(5)
```

```python
assert (df_soil_r.columns == df_veg_r.columns).all()
```
