```python
from greensight.utils import DIR_DATA
import pandas as pd
```

```python
# load DataFrame
s1_path = DIR_DATA / "processed_data/sentinel_one_df.hdf"

s1_data = pd.read_hdf(s1_path, key="df")
s1_data.head(5)
```

```python
# load weights DataFrame
weights_path_1 = DIR_DATA / "processed_data/sentinel_one_weights.hdf"

df_weights_1 = pd.read_hdf(weights_path_1, key="df")
df_weights_1.head(5)
```

```python
# weights df shape must equal data shape
assert df_weights_1.shape == s1_data.shape
```

```python
# load DataFrame
s2_path = DIR_DATA / "processed_data/sentinel_two_df.hdf"

s2_data = pd.read_hdf(s2_path, key="df")
s2_data.head(5)
```

```python
# load weights DataFrame
weights_path_2 = DIR_DATA / "processed_data/sentinel_two_weights.hdf"

df_weights_2 = pd.read_hdf(weights_path_2, key="df")
df_weights_2.head(5)
```

```python

# load DataFrame
si_path = DIR_DATA / "processed_data/sentinel_two_indices_df_droppedna.hdf"

si_data = pd.read_parquet(si_path)
si_data.head(5)
```

```python
# load weights DataFrame
weights_path_i = DIR_DATA / "processed_data/sentinel_two_index_weights_droppedna.hdf"

df_weights_i = pd.read_parquet(weights_path_i)
df_weights_i.head(5)
```

```python
# combine
s1_combined = s1_data*df_weights_1
s1_combined = s1_combined.groupby(level=["greenbelt", "band"], axis=1).sum()
s1_combined.columns = pd.MultiIndex.from_tuples([(i[0].replace("-", " ").replace(",", ""), i[1]) for i in s1_combined.columns])
s1_combined.columns.names = ["greenbelt", "band"]
s1_combined.head(5)
```

```python
# combine
s2_combined = s2_data*df_weights_2
s2_combined = s2_combined.groupby(level=["greenbelt", "band"], axis=1).sum(min_count=1)
s2_combined.columns = pd.MultiIndex.from_tuples([(i[0].replace("-", " ").replace(",", ""), i[1]) for i in s2_combined.columns])
s2_combined.columns.names = ["greenbelt", "band"]
s2_combined.head(5)
```

```python
# combine
si_combined = si_data*df_weights_i
si_combined = si_combined.groupby(level=["greenbelt", "band"], axis=1).sum()
si_combined.columns = pd.MultiIndex.from_tuples([(i[0].replace("-", " ").replace(",", ""), i[1]) for i in si_combined.columns])
si_combined.columns.names = ["greenbelt", "band"]
si_combined.head(5)
```

```python
# load JULES data
path_soil = DIR_DATA / "processed_data/jules_soil.hdf"
path_veg = DIR_DATA / "processed_data/jules_veg.hdf"
```

```python
jules_soil = pd.read_hdf(path_soil, key="df")
jules_soil = jules_soil.shift(-14, freq='d')
jules_soil = jules_soil.resample("MS").mean()
jules_soil.columns.name = "greenbelt"
jules_soil.columns = [i.replace("-", " ").replace(",", "") for i in jules_soil.columns]
jules_soil
```

```python
jules_veg = pd.read_hdf(path_veg, key="df")
jules_veg = jules_veg.shift(-14, freq='d')
jules_veg = jules_veg.resample("MS").mean()
jules_veg.columns.name = "greenbelt"
jules_veg.columns = [i.replace("-", " ").replace(",", "") for i in jules_veg.columns]
jules_veg
```

```python
jules_soil.columns = pd.MultiIndex.from_product([jules_soil.columns, ["soil"]], names=["greenbelt", "band"])
jules_soil
```

```python

jules_veg.columns = pd.MultiIndex.from_product([jules_veg.columns, ["vegetation"]], names=["greenbelt", "band"])
jules_veg.head(5)
```

```python
df_all = pd.concat([jules_soil, jules_veg, s1_combined, s2_combined, si_combined], axis=1, join="inner")
```

```python
# df_all = df_all.groupby(level=0, axis=1).apply(
#     lambda df: df.droplevel(0, axis=1).T.drop_duplicates().T
# )

df_all = df_all.groupby(level=0, axis=1).apply(
    lambda df: df.droplevel(0, axis=1)
    )
```

```python
# path_out = DIR_DATA / "processed_data/data_all.hdf"

path_out = DIR_DATA / "processed_data/data_all_droppedna.hdf"
df_all = df_all.dropna()
df_all.to_parquet(path_out)
```
